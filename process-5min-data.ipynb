```python
import os
import pandas as pd
import numpy as np
import shutil
```

```python
data = "/content/drive/MyDrive/cs410/data" # path to data dir
files_to_split = os.path.join(data, "subject_8") # or whatever path to your two files
```

```python
files = os.listdir(files_to_split)
files
```

```python
eyes_closed = pd.read_csv(os.path.join(files_to_split, files[2]))
eyes_open = pd.read_csv(os.path.join(files_to_split, files[0]))
```

```python
eyes_open.head()
```

```python
mod = eyes_open.shape[0] % 3100 #remainder of dividing df length by per-split length (3100)
# this splits into 10 datasets with 3100 rows (~30 seconds)
# can adjust the modulus to make more/less datasets (the smaller mod value, the better)
start = eyes_open.shape[0]-mod
end = eyes_open.shape[0]
eyes_open_trimmed = eyes_open.drop([i for i in range(start,end)])
eyes_open_trimmed.shape[0] % 3100
```

```python
eo1, eo2, eo3, eo4, eo5, eo6, eo7, eo8, eo9, eo10 = np.split(eyes_open_trimmed, 10)
```
